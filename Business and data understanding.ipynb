# Data understanding

<p><h3>Data Source</h3></p>

<p style='font-size:17px'>For this work we will use the Washington State Department of Transportation Annual Collision Data Summary Reports to obtain information about Seattle’s Collision. This Dataset provide all the information about collision in the streets of Seattle from 2004 to today.<br>
All the collisions are given a code [SEVERITY CODE] which corresponds to the severity of the collision, the date of the accident also the conditions in that day. We can pinpoint where the collision happened via the latitude and longitude while having the possibility to assess the party involves. Additionally, a description of the collision is added.</p>





<p><h3>-How we will use the data</h3></p>
<p style='font-size:17px'>From the Data set we will isolate all the accidents that involve a car and a unicycle, bicycle, tricycle. We will try to find the relation between the collisions with light, weather, and road conditions. Also we will look at the location of the different accident and search for all possible correlations.</p>
